```python
data_2019 = data.loc[data["month"] == '2019_06']
```

```python
data_2019_pivot = pd.pivot_table(data = data_2019,
               index = 'neighbourhood',
               values = 'id',
               aggfunc = 'count').reset_index().sort_values("id", ascending = False).head(10)
ax = data_2019_pivot.plot(x='neighbourhood',kind='bar', width=0.8, figsize=(10, 6), color=['blue'])

plt.legend().set_visible(False)
ax.set_xlabel('Neighborhood')
ax.set_ylabel('Number of Reviews')
ax.set_title('Number of Reviews in June 2019 in top 10 neighbourhoods')

plt.show()
```

```python
data_2020 = data.loc[data["month"] == '2020_06']
```

```python
data_2020_pivot = pd.pivot_table(data = data_2020,
               index = 'neighbourhood',
               values = 'id',
               aggfunc = 'count').reset_index().sort_values("id", ascending = False).head(10)
ax = data_2020_pivot.plot(x='neighbourhood',kind='bar', width=0.8, figsize=(10, 6), color=['blue'])

plt.legend().set_visible(False)
ax.set_xlabel('Neighborhood')
ax.set_ylabel('Number of Reviews')
ax.set_title('Number of Reviews in June 2020 in top 10 neighbourhoods')

plt.show()
```

```python
df_filtered = data[data['month'].isin(['2019_06', '2020_06'])]

df_aggregated = df_filtered[['neighbourhood', 'month', 'number_of_reviews']].groupby(['neighbourhood', 'month']).sum().reset_index()
df_pivot = df_aggregated.pivot(index='neighbourhood', columns='month', values='number_of_reviews')

ax = df_pivot.plot(kind='bar', width=0.8, figsize=(10, 6), color=['blue', 'orange'])

ax.set_xlabel('Neighborhood')
ax.set_ylabel('Number of Reviews')
ax.set_title('Number of Reviews in June 2019 and June 2020 by Neighborhood')
ax.legend(title='Month')

plt.show()
```
